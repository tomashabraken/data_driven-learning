```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('iris.csv')
df.head()
```

```python
sns.scatterplot(x='PetalLengthCm', y='SepalLengthCm', hue='Species', data=df)
```

What possible clusters do you see with the naked eye? How do they relate to the three species?

I see 2 possible clusters with the naked eye. One left bottom(Iris-setosa) and the other one is the bigger one, this one contains two species(Iris-versicolor and Iris-virginica). 


```python
km = KMeans(n_clusters=3) #create a new k-means model with 3 clusters
X = df[['SepalLengthCm', 'PetalLengthCm']] #get the X variables from the dataframe
km = km.fit(X) #calculate the cluster centers
df['cluster'] = km.predict(X) #predict the clusters of each observation and store in the dataframe
df.head()
```

```python
sns.scatterplot(x='PetalLengthCm', y='SepalLengthCm', hue='cluster', data=df)
```
