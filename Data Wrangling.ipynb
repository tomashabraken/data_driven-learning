```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_movies = pd.read_csv("movies.csv")
df_movies.head()
```

```python
df_ratings = pd.read_csv("ratings.csv")
df_ratings.head()
```

```python
df = pd.merge(df_movies, df_ratings, on='movieId') 
df.head()
```

```python
df_piv = pd.pivot_table(data=df, columns='title', values='rating', index = 'userId')
df_piv.head()
```

```python
sns.distplot(df_piv['Toy Story (1995)'].dropna(), kde=False, bins=10)
```
